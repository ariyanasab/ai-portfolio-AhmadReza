```python
import pandas as pd

# Create sample data
data = {
    'Name': ['Ali', 'Sara', 'Reza', 'Maryam', 'Ahmad'],
    'Score': [75, 88, 92, 60, 83]
}

df = pd.DataFrame(data)

# Add a normalized score column
df['Normalized'] = (df['Score'] - df['Score'].min()) / (df['Score'].max() - df['Score'].min())

df
```
